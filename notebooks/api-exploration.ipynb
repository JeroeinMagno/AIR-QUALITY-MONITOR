```python
import json

from openaq import OpenAQ
```

```python
with open("../secrets.json") as f:
    secrets = json.load(f)
    f.close()

    openaq_key = secrets["openaq-api-key"]
```

```python
client = OpenAQ(api_key=openaq_key)

client
```

```python
ph_bbox_search_locations = client.locations.list(
    bbox=(115.507840,5.674296,130.182708,20.152223),
    limit = 1000
)

ph_bbox_search_locations.meta
```

```python
ph_bbox_search_locations.results[0].__dict__
```

```python
locations_info = {}

for location in ph_bbox_search_locations.results:

    locations_info[location.id] = location.name

locations_info
```

```python
len(locations_info)
```

```python
with open("../location.json", "w", encoding="utf-8") as f:
    json.dump(locations_info, f, ensure_ascii=False, indent=4)
    f.close()
```
